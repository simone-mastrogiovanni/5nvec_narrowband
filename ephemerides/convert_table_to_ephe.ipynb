```python
import pandas as pd
import numpy as np
from astropy import units as u
from astropy.coordinates import Angle
```

```python
df = pd.read_csv('O3_HI_list.txt',sep='\t')
```

```python
df
```

```python
tot_pulsar = len(df)

for i in range(tot_pulsar):
    f = open('P'+str(i)+'.par', 'w')
    
    name = 'P'+str(i)
    f0 = df['f0'][i]/2
    f1 = df['fdot'][i]/2
    dec = Angle(df['dec'][i],u.radian)
    ra = Angle(df['ra'][i],u.radian)
    
    f.write('PSRJ\tP{:d}\n'.format(i))
    f.write('NAME\tP{:d}\n'.format(i))
    f.write('RAJ\t{:f}:{:f}:{:f}\n'.format(ra.hms[0],ra.hms[1],ra.hms[2]))
    f.write('DECJ\t{:f}:{:f}:{:f}\n'.format(
        dec.signed_dms[0]*dec.signed_dms[1],dec.signed_dms[2],dec.signed_dms[3]))
    f.write('F0\t{:.9f}\n'.format(f0))
    f.write('F1\t{:.6e}\n'.format(f1))
    f.write('POSEPOCH\t{:d}\n'.format(58484))
    f.write('PEPOCH\t{:d}\n'.format(58484))
    
    f.close()

    
```
